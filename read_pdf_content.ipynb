```python
import requests
from bs4 import BeautifulSoup

url = "https://www.sipac.ufpi.br/public/jsp/processos/processo_detalhado.jsf?id=594067"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

# Find the outer table
table = soup.find('table', class_='formulario', width='100%')

# Extract data from each row (assuming consistent structure)
for row in table.find_all('tr')[1:]:  # Skip the header row
    data_cells = row.find_all('td', recursive=False)  # Avoid nested tables
    # Process each data cell here (e.g., print text content)
    for cell in data_cells:
        print(cell.text.strip())
```

```python
import os

def get_pdf_filenames(folder_path):
    """
    Gets the filenames of all PDF files in a given folder.

    Args:
        folder_path (str): The path to the folder containing the PDF files.

    Returns:
        list: A list of PDF filenames.
    """

    pdf_filenames = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.pdf'):
            pdf_filenames.append(filename)
    return pdf_filenames

def sort_pdf_filenames(filenames):
    """
    Sorts a list of PDF filenames numerically, following the specified order.

    Args:
        filenames (list): A list of PDF filenames.

    Returns:
        list: A sorted list of PDF filenames.
    """

    # Extract the numerical part from each filename
    numbers = [int(filename.split('d')[1].split('.')[0]) for filename in filenames]

    # Sort the numbers and filenames together
    sorted_pairs = sorted(zip(numbers, filenames))

    # Extract the sorted filenames
    sorted_filenames = [filename for _, filename in sorted_pairs]

    return sorted_filenames
```

```python
# Replace 'your_pdf_folder' with the actual path to your PDF folder
folder_path = 'pdfs'

pdf_files = get_pdf_filenames(folder_path)

print(pdf_files)

# Sort the filenames
sorted_filenames = sort_pdf_filenames(pdf_files)

# Print the sorted filenames
print(sorted_filenames)
```
